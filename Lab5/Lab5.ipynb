```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
```

```python
a=0.32
b=0.04
c=0.42
d=0.02
y0=[9, 20]
```

```python
def syst2(y, t):
    y1,y2=y
    return [-a*y1+b*y1*y2, c*y2-d*y1*y2]
```

```python
t=np.arange(0, 300, 0.1)
```

```python
y=odeint(syst2, y0, t)
```

```python
y11=y[:,0]
y21=y[:,1]
```

```python
fig=plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('1.png', dpi=600)
```

```python
fig=plt.figure(facecolor='white')
plt.plot(t, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('2.png', dpi=600)
```

```python
fig=plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
fig.savefig('3.png', dpi=600)
```
